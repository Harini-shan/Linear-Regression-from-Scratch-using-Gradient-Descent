**Import Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
```

**Create Dataset**

```python
X= np.array([1,2,3,4,5])
y= np.array([2,4,6,8,10])
```

**Initialize Parameters and Define Hyperparameters**

```python
 # y= m.x + b
# Initialize slope m and intercept b
m1=0.0
b1=0.0

# Hyperparameters
learning_rate = 0.001 # alpha
epochs = 1000 # No. of iterations
n = len(X)  # No.of samples
```

**Gradient Descent Training Loop (using (2/n) â€” the standard derivative of MSE)**

```python
for epoch in range(epochs):
    y_pred = m1 * X + b1
    error = y_pred - y

    # partial differenciation
    dm = (2/n) * np.dot(error,X)
    db = (2/n) * np.sum(error)

    # Updating parameters
    m1 -= learning_rate * dm
    b1 -= learning_rate * db

    # Print loss every 100 iterations
    if epoch % 100 == 0:
        loss = (1/n) * np.sum(error**2)
```

```python
print(f"\nFinal model: y = {m1:.2f}x + {b1:.2f}")
```

```python
# Plot
plt.scatter(X, y, label='Actual')
plt.plot(X, m1*X + b1, color='red', label='Predicted_1')
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression using Gradient Descent")
plt.show()
```

**Gradient Descent Training Loop (using (-1/n) â€” from the (1/2n) version of MSE)**

```python
# y= m.x + b
# Initialize slope m and intercept b
m2=0.0
b2=0.0

# Hyperparameters
learning_rate = 0.001 # alpha
epochs = 1000 # No. of iterations
n = len(X)  # No.of samples
```

```python
for epoch in range(epochs):
    y_pred = m2 * X + b2
    error = y - y_pred

    # partial differenciation
    dm = (-1/n) * np.dot(error,X)
    db = (-1/n) * np.sum(error)

    # Updating parameters
    m2 -= learning_rate * dm
    b2 -= learning_rate * db
```

```python
# Plot
plt.scatter(X, y, label='Actual')
plt.plot(X, m2*X + b2, color='blue', label='Predicted_2')
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Linear Regression using Gradient Descent")
plt.show()
```

```python
plt.plot(X, m1*X + b1, color='red', label='Predicted (2/n gradient)')
plt.plot(X, m2*X + b2, color='blue', linestyle='--', label='Predicted (-1/n gradient)')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Comparison of Linear Regression Predictions")
plt.legend()
plt.grid(True)
plt.show()
```
